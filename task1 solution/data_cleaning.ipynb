```python
import pandas as pd
import numpy as np
```

```python
#import csv as dataframe
camp=pd.read_csv('camp_data.csv')
biz=pd.read_csv('business_vertical.csv')
channel=pd.read_csv('channel_name.csv')
country=pd.read_csv('country.csv')
```

```python
#check camp missing value
camp.isnull().sum(axis=0)
```

```python
list(camp['goal_type'].unique())
```

```python
list(camp['clicks'].unique())
```

```python
list(camp['conversions'].unique())
```

```python
list(camp['impressions'].unique())
```

```python
#fill in camp missing data

#fill column 'goal_type' nan values with 0, which means this record doesn't choose any goal type
camp['goal_type']=camp['goal_type'].fillna(0)

#fill column 'conversions' nan values with 0, reason: all other value in conversions column is 0
camp['conversions']=camp['conversions'].fillna(0)
```

```python
#drop camp missing data

#remove rows where column 'impressions' and 'clicks' are empty
camp.dropna(subset=['impressions'],inplace=True)
camp.dropna(subset=['clicks'],inplace=True)
```

```python
camp.isnull().sum(axis=0)
```

```python
len(camp)
```

```python
#join camp with other three dataframe
campnew=pd.merge(camp, biz, left_on="business_vertical", right_on="label").drop('label', axis=1)
campnew=pd.merge(campnew, channel, left_on="channel_name", right_on="label").drop('label', axis=1)
campnew=pd.merge(campnew, country, left_on="country", right_on="label").drop('label', axis=1)
```

```python
campnew.isnull().sum(axis=0)
```

```python
len(campnew)
```

```python
campnew.to_csv('.\clean_data\campnew.csv')
```
